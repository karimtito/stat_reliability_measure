# Analysing linear tests results

## 1. Imports

```python
import pandas as pd
```

## 2. Analysis of langevin methods

```python
agg_res=pd.read_csv('../logs/linear_tests/aggr_res.csv')
agg_res2=pd.read_csv('../logs/linear_tests/aggr_res2.csv')
agg_res3=pd.read_csv('../logs/linear_tests/aggr_res3.csv')
T_sel=(agg_res['T']==5).to_numpy()
df_sel=agg_res
#df_sel=df_sel[T_sel]
ad_sel=(df_sel['method']=='langevin_adapt').to_numpy()
df_sel=df_sel[~ad_sel]
alpha_sel=(df_sel['alpha']==0.02).to_numpy()
df_sel=df_sel[~alpha_sel]
by_alpha=df_sel.groupby(['method','p_t','alpha','T','N','rho'])
```

```python
by_alpha[['mean rel error','n_rep','mean abs error','mean time','bias','std est','mean est','freq underest']].mean()
```

```python
T_sel=(agg_res3['T']==5).to_numpy()
df_sel=agg_res3
#df_sel=df_sel[T_sel]
ad_sel=(df_sel['method']=='langevin_adapt').to_numpy()
df_sel=df_sel[~ad_sel]
alpha_sel=(df_sel['alpha']==0.02).to_numpy()
df_sel=df_sel[~alpha_sel]
pt_sel=df_sel['p_t']==1e-15
df_sel=df_sel[pt_sel]
by_alpha=df_sel.groupby(['method','p_t','alpha','T','N'])
```

```python
by_alpha[['mean rel error','n_rep','mean abs error','bias','std est','mean est','mean time']].mean()
```
